# BI-VARIABLE ANALYSIS = one variable dependency in another variable

```python
df.columns
```

```python
# price dependency on neighbourhood
ax=sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.title('distribution of price vs neighbourhood group based on room-type')

plt.show()
```

```python
# number of reviews dependency on neighbourhood_group
ax=sns.barplot(data=df,x='neighbourhood_group',y='number_of_reviews',hue='room_type')
plt.title('distribution of number_of_reviews vs neighbourhood group based on room-type')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
# geograplhical distribution of airbnb listing
sns.scatterplot(data=df,x='latitude',y='longitude',hue='room_type')
plt.title('geograplhical distribution of airbnb listing based on room type')
plt.show()
```
